```python
import gradio as gr
from openai import OpenAI


API_KEY = "API_Key"

client = OpenAI(
    api_key=API_KEY,
    base_url="https://api.z.ai/api/paas/v4/"
)

# ==========================================
# SYSTEM PROMPT
# ==========================================

SYSTEM_PROMPT = """
You are InnerEcho, an emotional wellness companion.

You are:
- empathetic
- calm
- conversational
- emotionally supportive

Do not sound robotic.
Respond naturally like ChatGPT.
"""

# ==========================================
# CHAT FUNCTION
# ==========================================

def chat_with_innerecho(message, history):

    if history is None:
        history = []

    messages = [
        {
            "role": "system",
            "content": SYSTEM_PROMPT
        }
    ]

    # Add previous chat history
    for item in history:

        if item["role"] == "user":
            messages.append(
                {
                    "role": "user",
                    "content": item["content"]
                }
            )

        elif item["role"] == "assistant":
            messages.append(
                {
                    "role": "assistant",
                    "content": item["content"]
                }
            )

    # Add current user message
    messages.append(
        {
            "role": "user",
            "content": message
        }
    )

    try:

        response = client.chat.completions.create(
            model="glm-4.5-flash",
            messages=messages,
            temperature=0.7,
            max_tokens=200
        )

        bot_reply = response.choices[0].message.content

        if bot_reply is None:
            bot_reply = "I'm here with you."

    except Exception as e:

        bot_reply = f"API Error: {str(e)}"

    # Update history
    history.append(
        {
            "role": "user",
            "content": message
        }
    )

    history.append(
        {
            "role": "assistant",
            "content": bot_reply
        }
    )

    return "", history


# ==========================================
# CLEAR FUNCTION
# ==========================================

def clear_chat():
    return []


# ==========================================
# INTERFACE
# ==========================================

with gr.Blocks(theme=gr.themes.Soft()) as demo:

    gr.Markdown("# 🌙 InnerEcho: Emotional Wellness Companion")

    gr.Markdown(
        "You are important.\n\nI'm really glad you're here :)"
    )

    chatbot = gr.Chatbot(
        type="messages",
        height=500
    )

    msg = gr.Textbox(
        placeholder="Talk to InnerEcho...",
        label="Message"
    )

    with gr.Row():

        send_btn = gr.Button("Send")

        clear_btn = gr.Button("Clear Chat")

    # Send button
    send_btn.click(
        chat_with_innerecho,
        inputs=[msg, chatbot],
        outputs=[msg, chatbot]
    )

    # Enter key
    msg.submit(
        chat_with_innerecho,
        inputs=[msg, chatbot],
        outputs=[msg, chatbot]
    )

    # Clear chat
    clear_btn.click(
        clear_chat,
        outputs=chatbot
    )

# ==========================================
# LAUNCH
# ==========================================

demo.launch(share=True)
```
